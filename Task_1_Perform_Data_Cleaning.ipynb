```python
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
import csv
```

```python
data=pd.read_csv("C:/Users/mohan/Desktop/Data analytics Internship/dataset/test.csv")
data
```

```python
data.isnull().sum()
```

```python

data['Age'] = data['Age'].fillna(data['Age'].median())


data.dropna(subset=['Fare'], inplace=True)

data.drop(columns=['Cabin'], inplace=True)

print(data.isnull().sum())
```

```python
data['PassengerId'].value_counts().plot(kind='box')
plt.title('Boxplot of PassengerId')
plt.show()
data['Pclass'].value_counts().plot(kind='box')
plt.title('Boxplot of Pclass')
plt.show()
data['Sex'].value_counts().plot(kind='box')
plt.title('Boxplot of Sex')
plt.show()
data['Age'].value_counts().plot(kind='box')
plt.title('Boxplot of Age')
plt.show()
data['Name'].value_counts().plot(kind='box')
plt.title('Boxplot of Name')
plt.show()
data['SibSp'].value_counts().plot(kind='box')
plt.title('Boxplot of SibSp')
plt.show()
data['Parch'].value_counts().plot(kind='box')
plt.title('Boxplot of Parch')
plt.show()
data['Ticket'].value_counts().plot(kind='box')
plt.title('Boxplot of Ticket')
plt.show()
data['Fare'].value_counts().plot(kind='box')
plt.title('Boxplot of Fare')
plt.show()
data['Embarked'].value_counts().plot(kind='box')
plt.title('Boxplot of Embarked')
plt.show()
```

```python
Q1 = data['Fare'].quantile(0.25)
Q3 = data['Fare' ].quantile(0.75)
IQR = Q3-Q1
lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR
data = data[(data['Fare'] >= lower_bound) & (data['Fare'] <= upper_bound)]

print(data)
```
