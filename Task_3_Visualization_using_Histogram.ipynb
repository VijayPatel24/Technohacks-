<h3>Task 3:Visualization using Histogram

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("C:/Users/mohan/Desktop/Data analytics Internship/dataset/iris.csv")
data
```

```python
data.dtypes
```

```python
plt.hist(data['sepal_length'], bins=10, edgecolor='k', alpha=0.7)
plt.xlabel('Sepal Length Cm')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```

```python
plt.hist(data['sepal_width'], bins=10, edgecolor='k', alpha=0.7)
plt.xlabel('Sepal Width Cm')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Width')
plt.show()
```

```python

plt.hist(data['petal_length'], bins=10, edgecolor='k', alpha=0.7)
plt.xlabel('Petal lenght Cm')
plt.ylabel('Frequency')
plt.title('Histogram of Petal lenght')
plt.show()
```

```python
plt.hist(data['petal_width'], bins=10, edgecolor='k', alpha=0.7)
plt.xlabel('Petal Width Cm')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Width')
plt.show()
```

```python
species_count = data['species'].value_counts()
plt.bar(species_count.index, species_count.values)
plt.xlabel('Species')
plt.ylabel('Frequency')
plt.title('Distribution of Species')
plt.show()
```
